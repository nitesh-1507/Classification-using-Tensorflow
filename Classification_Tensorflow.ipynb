# Classification using Tensorflow

```python
# importing packages

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow as tf
tf.logging.set_verbosity(tf.logging.ERROR)
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing data-set

data = pd.read_csv('iris.csv')
```

```python
# display of data

data[1:20]
```

```python
# Creating features and target attribute

X_features=['sepal_length','sepal_width','petal_length','petal_width']
Y_label=['species']
```

```python
# Creting features and target matrix and one-hot encoding

X=data[X_features]
Y=data[Y_label]

encoder=LabelEncoder()
y1=encoder.fit_transform(Y)
Y=pd.get_dummies(y1).values
```

```python
## One-hot encoder assignment

# sentosa=[1,0,0]
# versicolor = [0,1,0]
# verginica = [0,0,1]

# Display
Y
```

```python
# Train , test split using sklearn

train_X, test_X, train_Y, test_Y=train_test_split(X,Y, test_size=0.2, random_state=0)
```

```python
# Layers creating using keras

l0 = tf.keras.layers.Dense(units=4, input_shape=(4,), activation='relu')
l1 = tf.keras.layers.Dense(units=3, activation='softmax')
model = tf.keras.Sequential([l0,l1])
```

```python
# Model compiling

model.compile(tf.keras.optimizers.Adam(0.02), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# Fitting model using 100 iterations

model.fit(train_X, train_Y, epochs=100)
```

```python
# Storing test predictions
test_pred=model.predict(test_X)
test_pred
```

```python
# Creting list of original class and predicted class

y_test_class=np.argmax(test_Y, axis=1)
y_pred_class=np.argmax(test_pred, axis=1)
```

```python
# Classification report and Confusion matrix

print(classification_report(y_test_class, y_pred_class))
print(confusion_matrix(y_test_class, y_pred_class))
```
